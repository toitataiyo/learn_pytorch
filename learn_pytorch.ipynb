# 線形回帰 (自分の手で1から行う)

```python
import torch

w_true = torch.tensor([1, 2, 3.])
X = torch.cat([torch.ones(10, 1), torch.randn(10, 2)], 1)
print(X)
y = torch.mv(X, w_true) + torch.randn(10) * 0.5 #w_trueをかけたもの
print(y)
w = torch.randn(3, requires_grad=True) #重みが最適化の対象, こいつでベクトルを行う
gamma = 0.1
```

```python
losses = []
y_preds = []

'''
yはX([1, ガウス分布に従う数x1, ガウス分布に従う数x2])とw_true([1, 2, 3]))の積を10個
つまり y = 1 + 2x1 + 3x2
y_predは、yで用いたXと同じXとw([適当な重みw1, 適当な重みw2, 適当な重みw3])の積を10個
つまり y = w1 + w2x2 + w3x3

loss = y - y_pred
yはXとwの積であり、wは微分可能
lossを最小化するためにwで微分して勾配法によりデータを更新していく
'''
for epoc in range(10):
    w.grad = None
    print(f"epoc:{epoc}")
    y_pred = torch.mv(X, w) #wは適当な重みで微分可能
    #print(f"y_pred:{y_pred}")
    loss = torch.mean((y-y_pred)**2)
    loss.backward()
    print(f"更新前の重み:{w}")
    print(f"微分によって得られた勾配:{w.grad}")
    w.data = w.data - gamma * w.grad.data #gamma = 0.1
    print(f"更新後の重み:{w}")

    losses.append(loss.item())
```

```python
from matplotlib import pyplot as plt
plt.plot(losses)
```

```python
print(w) #最適化された重み
```

# ロジスティック回帰によるirisのデータ分類

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data[:100]
y = iris.target[:100]
print(X) #1つのデータにつき4つの花の特徴量
print(y) #花の分類

#tensor型に直す
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
```

```python
net = nn.Linear(4, 1) #入力ノード数が4, 出力ノード数が1
loss_fn = nn.BCEWithLogitsLoss() #クロスエントロピー誤差の損失関数
optimizer = optim.SGD(net.parameters(), lr=0.25)
```

```python
losses = []

for epoc in range(100):
    optimizer.zero_grad()
    y_pred = net(X)
    loss = loss_fn(y_pred.view_as(y), y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
```

```python
plt.plot(losses)
```

```python
h = net(X)
prob = nn.functional.sigmoid(h)
y_pred = prob > 0.5
(y.byte() == y_pred.view_as(y)).sum().item()
```

# 手書き数字のデータのクラス分類(ロジスティック)

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
X = digits.data
y = digits.target

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.int64)

net = nn.Linear(X.size()[1], 10)
loss_fn = nn.CrossEntropyLoss() #irisの時に用いたクロスエントロピーの多クラス分類バージョン
optimizer = optim.SGD(net.parameters(), lr=0.02)
```

```python
print(X.size()[0])
print(X.size()[1])
```

```python
losses = []

for epoch in range(100):
    optimizer.zero_grad()
    y_pred = net(X)
    loss = loss_fn(y_pred, y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
```

```python
_, y_pred = torch.max(net(X), 1)
(y_pred == y).sum().item()/len(y)
```
